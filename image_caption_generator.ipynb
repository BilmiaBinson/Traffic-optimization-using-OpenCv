```python
# step 2 - import the libraries
import torch
from transformers import BlipProcessor , BlipForConditionalGeneration
from PIL import Image
import gradio as gr
```

```python
# step 3 - loading the functions from libraries
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
# step 4 - creating a python function called generate_caption
def generate_caption(image):
  # Process the image for the model
  rgb_image = image.convert('RGB')
  inputs = processor(images=rgb_image, return_tensors="pt")

  # Generate a caption
  out = model.generate(**inputs)

  # Decode the generated caption
  caption = processor.decode(out[0], skip_special_tokens=True)

  # Return the generated caption
  return caption
```

```python
# step 5 - building the interface
image_caption_app = gr.Interface(
    fn=generate_caption,
    inputs = gr.Image(type="pil",label="Upload an Image"),
    outputs = gr.Textbox(label="Generated Caption",lines=2),
    title="Caramel AI - Image Caption Generator",
    description="Please Upload an Image and Caramel AI will generate the caption",
    allow_flagging="never"
)
if __name__ == "__main__":
  image_caption_app.launch()
```
